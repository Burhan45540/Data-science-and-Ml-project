## üìÅ CSV vs JSON ‚Äî Complete Guide

---

### üîπ 1. What Is CSV?

**CSV (Comma-Separated Values)** is a simple **text-based format** to store tabular data.
Each line = one row. Values are separated by **commas**.

#### ‚úÖ Example CSV:

```
name,age,city
Alice,25,New York
Bob,30,London
```

‚û°Ô∏è Represents a table like:

| name  | age | city     |
| ----- | --- | -------- |
| Alice | 25  | New York |
| Bob   | 30  | London   |

---

### üîπ 2. What Is JSON?

**JSON (JavaScript Object Notation)** is a **structured data format**.
It supports key-value pairs and **nested objects**.

#### ‚úÖ Example JSON:

```json
[
  {
    "name": "Alice",
    "age": 25,
    "city": "New York"
  },
  {
    "name": "Bob",
    "age": 30,
    "city": "London"
  }
]
```

‚û°Ô∏è Same data as CSV but in **structured form**.

---

### üîÅ CSV vs JSON Comparison

| Feature       | CSV                    | JSON                             |
| ------------- | ---------------------- | -------------------------------- |
| Format        | Plain text, row/column | Text with nested key-value pairs |
| Structure     | Flat/tabular only      | Supports nested structures       |
| Readability   | Easy to read/write     | Slightly more complex            |
| File Size     | Smaller                | Slightly bigger                  |
| Use Cases     | Datasets, Excel, ML    | Web APIs, configs, nested data   |
| Parsing Speed | Faster (for flat data) | Slower (due to nesting)          |

## üêç Python Examples

---

### ‚úÖ 1. CSV Handling

#### üì• Read CSV:

```python
import pandas as pd

df = pd.read_csv("data.csv")
print(df)
```

#### üíæ Write CSV:

```python
df.to_csv("output.csv", index=False)
```

---

### ‚úÖ 2. JSON Handling

#### üì• Read JSON:

```python
import pandas as pd

df = pd.read_json("data.json")
print(df)
```

#### üíæ Write JSON:

```python
df.to_json("output.json", orient='records', indent=2)
```

---

## üîö Summary

| Task              | CSV             | JSON                        |
| ----------------- | --------------- | --------------------------- |
| Best for          | Flat table data | Nested or hierarchical data |
| Human readable    | ‚úÖ               | ‚úÖ (more syntax)             |
| Python read func  | `pd.read_csv()` | `pd.read_json()`            |
| Python write func | `df.to_csv()`   | `df.to_json()`              |

## ü•í What is `pickle` in Python?

**`pickle`** is a built-in Python module used to **serialize** and **deserialize** Python objects.

---

### üîπ Terms:

| Term        | Meaning                                          |
| ----------- | ------------------------------------------------ |
| Serialize   | Convert Python object ‚Üí byte stream (for saving) |
| Deserialize | Convert byte stream ‚Üí original Python object     |

---

## ‚úÖ Why use `pickle`?

* Save Python data structures (like DataFrames, lists, dicts) to a file.
* Load them back **without losing data or structure**.
* Faster than saving to CSV or JSON (binary format).

---

## üêç Basic Example

### ‚ñ∂Ô∏è Save (Serialize) Python object to pickle file:

```python
import pickle

data = {"name": "Alice", "age": 25}

# Save to file
with open("data.pkl", "wb") as file:
    pickle.dump(data, file)
```

---

### ‚è™ Load (Deserialize) back:

```python
# Load from file
with open("data.pkl", "rb") as file:
    loaded_data = pickle.load(file)

print(loaded_data)
```

---

## üìä Using `pickle` with Pandas:

### ‚ñ∂Ô∏è Save a DataFrame:

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [25, 30]
})

df.to_pickle("df.pkl")
```

### ‚è™ Load the DataFrame:

```python
df2 = pd.read_pickle("df.pkl")
print(df2)
```

---

## üìå Summary:

| Format | Use Case                  | Supports Complex Data | Human-Readable | File Size |
| ------ | ------------------------- | --------------------- | -------------- | --------- |
| CSV    | Tables only               | ‚ùå                     | ‚úÖ              | Small     |
| JSON   | Nested dictionaries/lists | ‚úÖ                     | ‚úÖ              | Medium    |
| Pickle | Any Python object         | ‚úÖ                     | ‚ùå (binary)     | Smallest  |
