## Static Method (Short & Simple)

* A **static method** belongs to the class, **not objects**.
* It **does not use `self`**.
* You can call it **without creating an object**.
* Use `@staticmethod` decorator to define it.

---

### Code:

```python
import math

class MathOperations:
    @staticmethod
    def find_sqrt(num):
        return math.sqrt(num)
```

---

### Call without object:

```python
result = MathOperations.find_sqrt(25)  # No object needed
print(result)  # Output: 5.0
```

---

### Call with object (also works, but object not required):

```python
obj = MathOperations()
print(obj.find_sqrt(25))  # Output: 5.0
```

---

### Difference:

| With Object                             | Without Object (Static Method)                |
| --------------------------------------- | --------------------------------------------- |
| Create object: `obj = MathOperations()` | No object needed                              |
| Call method: `obj.find_sqrt(25)`        | Call directly: `MathOperations.find_sqrt(25)` |
| Uses memory for object                  | More efficient, no extra memory               |

---

### Key Point:

* `result` is **just the value returned** (like `5.0`), **not an object**.

### Using Properties and Setters — Why?

* Control access to private attributes
* Validate values before setting (e.g., no negative length)
* Prevent invalid or unwanted data in your objects
* Hide internal details from outside code
* Make your class safer and more reliable
* Allow changing internal implementation without changing how users interact with the class

---

### Without Properties and Setters — What Happens?

* Direct access to attributes without checks
* Can assign wrong or invalid values easily (e.g., negative length)
* No way to add validation or extra logic on get/set
* Risk of corrupt or inconsistent object state
* Harder to maintain or change internals later

---

**In short:**
**Properties + setters = safer, cleaner, and more maintainable code.**

### Assignment 11: Abstract Base Class

````python
from abc import ABC,abstractmethod
import math

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Circle(Shape):
  def __init__(self,radius):
    self.radius=radius
    

  def area(self):
    return math.pi* (self.radius **2)

class Square(Shape):
  def __init__(self,side):
    self.side=side

  def area(self):
    return self.side**2
  
c1=Circle(5)
c1.area()

s1=Square(4)
s1.area()
    
















# from abc import ABC, abstractmethod

# class Shape(ABC):
#     @abstractmethod
#     def area(self):
#         pass

# class Circle(Shape):
#     def __init__(self, radius):
#         self.radius = radius

#     def area(self):
#         return math.pi * self.radius ** 2

# class Square(Shape):
#     def __init__(self, side):
#         self.side = side

#     def area(self):
#         return self.side ** 2

# # Test
# # circle = Circle(5)
# # square = Square(4)
# # print(circle.area())  # 78.53981633974483
# # print(square.area())  # 16
````

### Assignment 14: Class with Context Manager

Create a class named `FileManager` that implements the context manager protocol to open and close a file. Use this class to read the contents of a file.

````python
class FileManager:

  def __init__(self,filename,mode):
    self.filename=filename
    self.mode=mode
    self.file=None

  def __enter__(self):
    self.file=open(self.filename,self.mode)
    return self.file
  
  def __exit__(self, exc_type, exc_val, exc_tb):
    if self.file:
      self.file.close()


with FileManager("data.txt",'r') as f:
  content=f.read()
  print(content)

  






















# class FileManager:
#     def __init__(self, filename, mode):
#         self.filename = filename
#         self.mode = mode

#     def __enter__(self):
#         self.file = open(self.filename, self.mode)
#         return self.file

#     def __exit__(self, exc_type, exc_value, traceback):
#         self.file.close()

# # Test
# # with FileManager('sample.txt', 'r') as file:
# #     content = file.read()
# #     print(content)
````
