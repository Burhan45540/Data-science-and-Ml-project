### âœ… Conditional Statements in Python (Simple Explanation)

Conditional statements are used to **make decisions** in Python. They help your program decide what to do **based on conditions** (like if something is true or false).

---

### ðŸ”¹ 1. `if` Statement

Used when you want to do something **only if a condition is true**.

```python
age = 18
if age >= 18:
    print("You can vote")
```

ðŸŸ¢ Output: `You can vote`

---

### ðŸ”¹ 2. `if-else` Statement

Used when you want to choose **between two options**.

```python
age = 16
if age >= 18:
    print("You can vote")
else:
    print("You cannot vote")
```

ðŸŸ¢ Output: `You cannot vote`

---

### ðŸ”¹ 3. `if-elif-else` Statement

Used when you have **multiple conditions**.

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

ðŸŸ¢ Output: `Grade B`

---

### ðŸ”¹ 4. Nested `if`

You can also use an `if` inside another `if`.

```python
age = 20
citizen = True

if age >= 18:
    if citizen:
        print("You can vote")
    else:
        print("Only citizens can vote")
```

ðŸŸ¢ Output: `You can vote`

---

### ðŸ”¹ 5. Short-hand `if` (One-liner)

Used for **very short conditions**.

```python
a = 10
b = 5

print("a is greater") if a > b else print("b is greater")
```

ðŸŸ¢ Output: `a is greater`
